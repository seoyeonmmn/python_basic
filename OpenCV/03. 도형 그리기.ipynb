## 3. 도형 그리기
원하는 다각형 그려서 이미지 및 영상에 위치 표시 가능

### 빈 스케치북 만들기

```python
import cv2
import numpy as np

# 세로 480 x 가로 640, 3 Channel (RGB) 에 해당하는 스케치북 만들기
img = np.zeros((480, 640, 3), dtype=np.uint8) # 모두 '0'으로 채움
# img[:] = (255, 255, 255) # 전체 공간을 흰 색으로 채우기 (괄호 안에 수치는 RGB 값, BGR 순서)
# print(img)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 일부 영역 색칠

```python
import cv2
import numpy as np

img = np.zeros((480, 640, 3), dtype=np.uint8)
img[100:200, 200:300] = (255, 255, 255)  # 영역 설정하여 흰 색(및 값 조절하여 원하는 색)으로 채우기
# [세로 영역, 가로 영역]

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 직선
직선의 종류 (line type)

1. cv2.LINE_4 : 상하좌우 4 방향으로 연결된 선
1. cv2.LINE_8 : 대각선을 포함한 8 방향으로 연결된 선 (기본값)
1. cv2.LINE_AA : 부드러운 선 (anti-aliasing)



출처: https://www.youtube.com/watch?v=XK3eU9egll8 나도코딩, 파이썬 코딩 무료 강의 (활용편6) - OpenCV 이미지 처리, 얼굴 인식을 통한 캐릭터 씌우기를 해보아요

```python
import cv2
import numpy as np

img = np.zeros((480, 640, 3), dtype=np.uint8)

COLOR = (0, 255, 255) # 색 정의 (BGR);Yellow
THICKNESS = 3 # 두께 정의

# 그릴 위치, 시작 점(x, y), 끝 점, 색깔, 두께, line_type
cv2.line(img, (50, 100), (400, 50), COLOR, THICKNESS, cv2.LINE_8)
cv2.line(img, (50, 200), (400, 150), COLOR, THICKNESS, cv2.LINE_4)
cv2.line(img, (50, 300), (400, 250), COLOR, THICKNESS, cv2.LINE_AA)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 원

```python
import cv2
import numpy as np

img = np.zeros((480, 640, 3), dtype=np.uint8)

COLOR = (255, 255, 0)
RADIUS = 50 
THICKNESS = 10 

# 그릴 위치, 원의 중심점, 반지름, 색깔, 두께, 선 종류
cv2.circle(img, (200, 100), RADIUS, COLOR, THICKNESS, cv2.LINE_AA) 
cv2.circle(img, (400, 100), RADIUS, COLOR, cv2.FILLED, cv2.LINE_AA) # filled circle

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 사각형

```python
import cv2
import numpy as np

img = np.zeros((480, 640, 3), dtype=np.uint8)

COLOR = (0, 255, 0) 
THICKNESS = 3 


# 그릴 위치, 왼쪽 위 좌표, 오른쪽 아래 좌표, 색깔, 두께
cv2.rectangle(img, (100, 100), (200, 200), COLOR, THICKNESS) 
cv2.rectangle(img, (300, 100), (400, 200), COLOR, cv2.FILLED)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 다각형

```python
img = np.zeros((480, 640, 3), dtype=np.uint8)

COLOR = (203, 192, 255) 
THICKNESS = 2

pts1 = np.array([[100, 100], [200, 100], [100, 200]])  # 삼각형1
pts2 = np.array([[200, 100], [300, 100], [300, 200]])  # 삼각형2


# 그릴 위치, 그릴 좌표들, 닫힘 여부, 색깔, 두께, 선 종류
cv2.polylines(img, [pts1, pts2], True, COLOR, THICKNESS, cv2.LINE_AA) # 그릴 좌표들을 리스트형으로 정의하여 여러개 지정할 수 있음.

pts3 = np.array([[[100, 300], [200, 300], [100, 400]], [[200, 300], [300, 300], [300, 400]]])
cv2.fillPoly(img, pts3, COLOR, cv2.LINE_AA) 

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
